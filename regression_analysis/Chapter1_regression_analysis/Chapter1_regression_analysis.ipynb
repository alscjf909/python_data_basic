```python
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
dataset['target'] = boston.target
```

### 데이터셋 출력을 위해 
from sklearn.datasets import load_boston

반드시 모든 가능한 데이터를 pandas 데이터 구조인 DataFrame으로 프레임화한다.
