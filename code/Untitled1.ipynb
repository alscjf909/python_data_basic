```python
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
dataset['target']=boston.target
```

```python
%matplotlib inline
scatter = dataset.plot(kind='scatter',x='RM', y='target')
```
